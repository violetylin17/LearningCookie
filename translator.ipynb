```python
from transformers import pipeline
```

```python
#get your input
input_request = input('Text: ')
```

```python
translator = pipeline(
    "translation",
    model="facebook/nllb-200-distilled-600M",
    src_lang="eng_Latn",
    tgt_lang="zho_Hant"
)
```

```python
result = translator(input_request)
print(result[0]["translation_text"])
```
